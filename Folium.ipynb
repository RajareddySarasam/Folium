```python
# lets locate India
india_map=folium.Map(location=[20.5937, 78.9629],
                    zoom_start=5)
india_map
```

```python
# Lets Try adding Marker to our India's Map

# Adding a marker at 'HYDERABAD'

india_map=folium.Map(location=[20.5937, 78.9629],
                    zoom_start=5)
folium.Marker(location=[17.4065, 78.4772],
                popup='Hyderabad').add_to(india_map)
india_map
```

```python
# Marking multiple Cities Once 

# we can also Add multiple markers without using MarkerCluster as well.
# But,Advantage of MarkerCluster is that it makes map'clear representation and not over crowding'

#cities-Mumbai,Delhir

from folium.plugins import MarkerCluster 
marker_cluster=MarkerCluster().add_to(india_map)
locations=[
    {'location':[28.679079,77.069710],'popup':'Delhi'},
    {'location':[19.076090,72.877426],'popup':'Mumbai'},
]
for loc in locations:
    folium.Marker(location=loc['location'],popup=loc['popup']).add_to(marker_cluster)
```
